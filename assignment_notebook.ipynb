````python
# ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Îî•Îü¨ÎãùÏö©)
sequence_length = 30

def create_sequences(X, y, seq_len=30):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_len):
        X_seq.append(X[i:i+seq_len])
        y_seq.append(y[i+seq_len])
    return np.array(X_seq), np.array(y_seq)

X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train.values, sequence_length)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val.values, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test.values, sequence_length)

print(f"ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ shape: {X_train_seq.shape}")
print(f"  - ÏÉòÌîå Ïàò: {X_train_seq.shape[0]}")
print(f"  - ÏãúÌÄÄÏä§ Í∏∏Ïù¥: {X_train_seq.shape[1]}")
print(f"  - ÌäπÏÑ± Ïàò: {X_train_seq.shape[2]}")
````

## 2. ÏòàÏ†ú: LSTM Î™®Îç∏ + ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à Ï†ÑÎûµ

Ïù¥ ÏòàÏ†úÎäî 2Í∞ÄÏßÄ ÌïµÏã¨ ÏöîÏÜåÎ°ú Íµ¨ÏÑ±Îê©ÎãàÎã§:

1. **ÏòàÏ∏° Î™®Îç∏**: Í∞ÄÍ≤© ÏÉÅÏäπ/ÌïòÎùΩÏùÑ ÏòàÏ∏°ÌïòÍ≥† ÌôïÎ•† Ï†úÍ≥µ
2. **Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ**: ÏòàÏ∏° ÌôïÎ•†Ïóê Îî∞Îùº Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à

### Ï†ÑÎûµ ÏòàÏãú
- ÏÉÅÏäπ ÌôïÎ•† 90% ‚Üí ÏûêÎ≥∏Ïùò 90% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 60% ‚Üí ÏûêÎ≥∏Ïùò 60% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 40% ‚Üí Ìà¨ÏûêÌïòÏßÄ ÏïäÏùå (ÌòÑÍ∏à Î≥¥Ïú†)

````python
# ÏòàÏ†ú Î™®Îç∏: LSTM Í∏∞Î∞ò 2-class Î∂ÑÎ•òÍ∏∞
class ExampleLSTMModel(nn.Module):
    """
    ÏòàÏ†ú: LSTM Í∏∞Î∞ò Í∞ÄÍ≤© Î∞©Ìñ• ÏòàÏ∏° Î™®Îç∏
    Ï∂úÎ†•: ÏÉÅÏäπ ÌôïÎ•† (0~1)
    """
    def __init__(self, input_size, hidden_size=64, dropout=0.2):
        super(ExampleLSTMModel, self).__init__()
        
        self.lstm1 = nn.LSTM(input_size, hidden_size, batch_first=True, num_layers=1)
        self.dropout1 = nn.Dropout(dropout)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        
        self.lstm2 = nn.LSTM(hidden_size, hidden_size//2, batch_first=True, num_layers=1)
        self.dropout2 = nn.Dropout(dropout)
        self.bn2 = nn.BatchNorm1d(hidden_size//2)
        
        self.fc1 = nn.Linear(hidden_size//2, 16)
        self.relu = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout)
        
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        lstm_out, _ = self.lstm1(x)
        lstm_out = self.dropout1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        lstm_out = self.bn1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        
        lstm_out, _ = self.lstm2(lstm_out)
        lstm_out = self.dropout2(lstm_out[:, -1, :])
        lstm_out = self.bn2(lstm_out)
        
        out = self.fc1(lstm_out)
        out = self.relu(out)
        out = self.dropout3(out)
        
        out = self.fc2(out)
        out = self.sigmoid(out)
        
        return out

# ÏòàÏ†ú Î™®Îç∏ ÏÉùÏÑ±
example_model = ExampleLSTMModel(
    input_size=X_train_seq.shape[2],
    hidden_size=64,
    dropout=0.2
).to(device)

print("ÏòàÏ†ú Î™®Îç∏ Íµ¨Ï°∞:")
print(example_model)
print(f"\nTotal parameters: {sum(p.numel() for p in example_model.parameters()):,}")
````

````python
# ÌïôÏäµ Ìï®Ïàò (Ï†úÍ≥µÎê®)
def train_model(model, train_loader, val_loader, epochs=100, lr=0.001, patience=15):
    model = model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y.unsqueeze(1))
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            predicted = (outputs > 0.5).float()
            train_total += batch_y.size(0)
            train_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_train_loss = train_loss / len(train_loader)
        train_acc = train_correct / train_total
        
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs, batch_y.unsqueeze(1))
                
                val_loss += loss.item()
                predicted = (outputs > 0.5).float()
                val_total += batch_y.size(0)
                val_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_val_loss = val_loss / len(val_loader)
        val_acc = val_correct / val_total
        
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}')
        
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break
    
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    
    return history

print("‚úÖ ÌïôÏäµ Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

````python
# ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ
print("ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏãúÏûë...")
example_history = train_model(
    model=example_model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=100,
    lr=0.001,
    patience=15
)
print("\n‚úÖ ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å!")
````

````python
# ÏòàÏ∏° Ìï®Ïàò (ÌôïÎ•† Ìè¨Ìï®)
def predict_with_probability(model, data_loader):
    """
    Î™®Îç∏ ÏòàÏ∏° Î∞è ÌôïÎ•† Î∞òÌôò
    """
    model.eval()
    predictions_prob = []
    
    with torch.no_grad():
        for batch_X, _ in data_loader:
            batch_X = batch_X.to(device)
            outputs = model(batch_X)
            predictions_prob.append(outputs.cpu().numpy())
    
    predictions_prob = np.vstack(predictions_prob).flatten()
    predictions = (predictions_prob > 0.5).astype(int)
    
    return predictions_prob, predictions

# ÏòàÏ†ú Î™®Îç∏ ÏòàÏ∏°
example_prob, example_pred = predict_with_probability(example_model, test_loader)

print(f"ÏòàÏ∏° ÏôÑÎ£å!")
print(f"ÏòàÏ∏° Ïàò: {len(example_pred)}")
print(f"ÏÉÅÏäπ ÏòàÏ∏°: {np.sum(example_pred == 1)}Í∞ú")
print(f"ÌïòÎùΩ ÏòàÏ∏°: {np.sum(example_pred == 0)}Í∞ú")
print(f"\nÌèâÍ∑† ÏÉÅÏäπ ÌôïÎ•†: {example_prob.mean():.2%}")
````

````python
# ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ
def simulate_probability_trading(predictions_prob, actual_prices, dates, 
                                 initial_capital=10000, transaction_fee=0.001,
                                 threshold=0.5, position_scaling=True):
    """
    ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò
    
    Parameters:
    -----------
    predictions_prob : array
        ÏÉÅÏäπ ÌôïÎ•† (0~1)
    actual_prices : array
        Ïã§Ï†ú Í∞ÄÍ≤©
    dates : array
        ÎÇ†Ïßú
    initial_capital : float
        Ï¥àÍ∏∞ ÏûêÎ≥∏
    transaction_fee : float
        Í±∞Îûò ÏàòÏàòÎ£å
    threshold : float
        Îß§Ïàò Í≤∞Ï†ï ÏûÑÍ≥ÑÍ∞í (Í∏∞Î≥∏ 0.5)
    position_scaling : bool
        True: ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à
        False: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
    """
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    trade_log = []
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                sell_value = btc_holdings * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL_ALL',
                    'price': current_price,
                    'prob': prob,
                    'amount': btc_holdings,
                    'value': btc_holdings * current_price,
                    'fee': btc_holdings * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings = 0
            continue
        
        # Ìà¨Ïûê ÎπÑÏú® Í≤∞Ï†ï
        if position_scaling:
            # ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à
            if prob > threshold:
                # ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê (Ïòà: 70% ÌôïÎ•† ‚Üí 70% Ìà¨Ïûê)
                invest_ratio = prob
            else:
                # ÏûÑÍ≥ÑÍ∞í ÎØ∏ÎßåÏù¥Î©¥ Îß§ÎèÑ ÎòêÎäî Í¥ÄÎßù
                invest_ratio = 0
        else:
            # Îã®Ïàú Ï†ÑÎûµ: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
            invest_ratio = 1.0 if prob > threshold else 0.0
        
        # ÌòÑÏû¨ Ìè¨ÏßÄÏÖò ÎπÑÏú®
        current_position_ratio = (btc_holdings * current_price) / portfolio_value if portfolio_value > 0 else 0
        
        # Ìè¨ÏßÄÏÖò Ï°∞Ï†ï
        target_btc_value = portfolio_value * invest_ratio
        current_btc_value = btc_holdings * current_price
        
        if target_btc_value > current_btc_value:  # Îß§Ïàò ÌïÑÏöî
            buy_cash = target_btc_value - current_btc_value
            if buy_cash > cash:
                buy_cash = cash
            if buy_cash > 0:
                buy_amount = (buy_cash * (1 - transaction_fee)) / current_price
                btc_holdings += buy_amount
                trade_log.append({
                    'date': dates[i],
                    'action': 'BUY',
                    'price': current_price,
                    'prob': prob,
                    'amount': buy_amount,
                    'value': buy_cash,
                    'fee': buy_cash * transaction_fee
                })
                cash -= buy_cash
        elif target_btc_value < current_btc_value:  # Îß§ÎèÑ ÌïÑÏöî
            sell_btc = (current_btc_value - target_btc_value) / current_price
            if sell_btc > btc_holdings:
                sell_btc = btc_holdings
            if sell_btc > 0:
                sell_value = sell_btc * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL',
                    'price': current_price,
                    'prob': prob,
                    'amount': sell_btc,
                    'value': sell_btc * current_price,
                    'fee': sell_btc * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings -= sell_btc
    
    final_value = portfolio_values[-1]
    total_return = (final_value - initial_capital) / initial_capital * 100
    
    total_trade_volume = sum(trade['value'] for trade in trade_log)
    total_fees_paid = sum(trade['fee'] for trade in trade_log)
    
    return {
        'initial_capital': initial_capital,
        'final_value': final_value,
        'total_return': total_return,
        'portfolio_values': portfolio_values,
        'trade_log': trade_log,
        'num_trades': len(trade_log),
        'total_trade_volume': total_trade_volume,
        'total_fees_paid': total_fees_paid
    }

print("‚úÖ Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

````python
# [Level 2] ÎßûÏ∂§Ìòï Ï†ÑÎûµ (Í≥µÍ≤©Ï†Å Ìà¨Ïûê)
print("\n=== Level 2: Custom Strategy Experiment ===\n")

# 1. ÌôïÎ•† Î∂ÑÌè¨ ÌôïÏù∏
print(f"Max Probability: {example_prob.max():.4f}")
print(f"Min Probability: {example_prob.min():.4f}")
print(f"Mean Probability: {example_prob.mean():.4f}")

# 2. ÏÉàÎ°úÏö¥ Ï†ÑÎûµ Ìï®Ïàò Ï†ïÏùò
def custom_strategy(predictions_prob, actual_prices, dates, initial_capital=10000, transaction_fee=0.001):
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    
    threshold = 0.5  # Í≥µÍ≤©Ï†ÅÏúºÎ°ú 0.5Î°ú ÎÇÆÏ∂§
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        # ÎßàÏßÄÎßâ ÎÇ† Îß§ÎèÑ
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                cash += btc_holdings * current_price * (1 - transaction_fee)
            continue
            
        # [ÏàòÏ†ïÎêú Î∂ÄÎ∂Ñ] Ìè¨ÏßÄÏÖò Ïä§ÏºÄÏùºÎßÅ Í≥µÏãù
        if prob > threshold:
            # ÏõêÎûòÎäî invest_ratio = prob ÏòÄÏßÄÎßå,
            # Ïó¨Í∏∞ÏÑúÎäî 0.5Î•º ÎÑòÎäî Ï†ïÎèÑÏóê 2Î•º Í≥±Ìï¥ÏÑú Îçî ÎßéÏù¥ Ìà¨ÏûêÌïòÍ≤å Ìï®
            # Ïòà: ÌôïÎ•† 0.6 -> (0.6 - 0.5) * 2 = 0.2 (20% Ìà¨Ïûê) -> ÎÑàÎ¨¥ Ï†ÅÏùå?
            # Ïòà: ÌôïÎ•† 0.6 -> 0.6 * 1.5 = 0.9 (90% Ìà¨Ïûê) -> Ïù¥Î†áÍ≤å Ìï¥Î¥ÖÏãúÎã§.
            invest_ratio = min(prob * 1.2, 1.0) # ÌôïÎ•†Ïùò 1.2Î∞∞ Ìà¨Ïûê (ÏµúÎåÄ 100%)
        else:
            invest_ratio = 0
            
        # Îß§Îß§ Î°úÏßÅ (Í∏∞Ï°¥Í≥º ÎèôÏùº)
        target_btc_value = portfolio_value * invest_ratio
        current_btc_value = btc_holdings * current_price
        
        if target_btc_value > current_btc_value:
            buy_val = target_btc_value - current_btc_value
            if buy_val > cash: buy_val = cash
            if buy_val > 0:
                amount = (buy_val * (1 - transaction_fee)) / current_price
                btc_holdings += amount
                cash -= buy_val
        elif target_btc_value < current_btc_value:
            sell_val = current_btc_value - target_btc_value
            amount = sell_val / current_price
            if amount > btc_holdings: amount = btc_holdings
            if amount > 0:
                cash_gained = amount * current_price * (1 - transaction_fee)
                btc_holdings -= amount
                cash += cash_gained

    total_return = (portfolio_values[-1] - initial_capital) / initial_capital * 100
    return total_return

# 3. Ïã§Ìñâ
lv2_return = custom_strategy(example_prob, test_prices_aligned, test_dates_aligned)
print(f"Level 2 Custom Strategy Return: {lv2_return:.2f}%")
````

````python
# [Level 3] 1. GRU Î™®Îç∏ Ï†ïÏùò Î∞è ÌïôÏäµ
import torch.nn as nn
import torch.optim as optim

class MyTradingModel(nn.Module):
    def __init__(self, input_size, hidden_size=128, num_layers=2, dropout=0.3):
        super(MyTradingModel, self).__init__()
        # LSTM ÎåÄÏã† GRU ÏÇ¨Ïö© (Íµ¨Ï°∞ Í∞úÏÑ†)
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        # Î∂ÑÎ•òÍ∏∞ ÏÑ±Îä• Í∞ïÌôîÎ•º ÏúÑÌïú FC Î†àÏù¥Ïñ¥ ÌôïÏû•
        self.fc = nn.Sequential(
            nn.Linear(hidden_size, 32),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        # x: (batch, seq_len, features)
        out, _ = self.gru(x)
        # ÎßàÏßÄÎßâ ÏãúÏ†ê(=Í∞ÄÏû• ÏµúÍ∑º Îç∞Ïù¥ÌÑ∞)Ïùò Ï∂úÎ†•Îßå ÏÇ¨Ïö©
        out = out[:, -1, :] 
        out = self.fc(out)
        return out

# Î™®Îç∏ ÏÉùÏÑ± (ÌååÎùºÎØ∏ÌÑ∞ Ï¶ùÍ∞ï)
my_model = MyTradingModel(
    input_size=X_train_seq.shape[2],
    hidden_size=128, 
    num_layers=2,
    dropout=0.3
).to(device)

print("=== Level 3: New GRU Model Created ===")
print(my_model)

# Î™®Îç∏ ÌïôÏäµ (Í∏∞Ï°¥ train_model Ìï®Ïàò Ïû¨ÏÇ¨Ïö©)
print("\nTraining New Model... (Please wait)")
my_history = train_model(
    model=my_model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=50,  # 50 ÏóêÌè≠ Ï†ïÎèÑÎ©¥ Ï∂©Î∂ÑÌï† ÎìØ Ìï©ÎãàÎã§
    lr=0.0005,  # ÌïôÏäµÎ•†ÏùÑ Ï°∞Í∏à ÎÇÆÏ∂∞ÏÑú ÏïàÏ†ïÏ†ÅÏúºÎ°ú
    patience=10
)
print("Training Complete!")
````

````python
# [Level 4] Î≥µÌï© Ï†ÑÎûµ (Model + RSI)
print("\n=== Level 4: Composite Strategy (Model + RSI) ===\n")

# 1. RSI Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
# test_start_idxÎäî ÏúÑÏóêÏÑú Ï†ïÏùòÎêú Î≥ÄÏàòÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§
rsi_values = btc_features['RSI_14'].iloc[test_start_idx:test_start_idx+len(my_prob)].values

# 2. Î≥µÌï© Ï†ÑÎûµ Ìï®Ïàò Ï†ïÏùò
def composite_strategy(predictions_prob, actual_prices, dates, rsi_values, initial_capital=10000, transaction_fee=0.001):
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    
    threshold = 0.5
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        rsi = rsi_values[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                cash += btc_holdings * current_price * (1 - transaction_fee)
            continue
            
        # 1Ï∞® ÌåêÎã®: Î™®Îç∏ ÌôïÎ•† Í∏∞Î∞ò
        if prob > threshold:
            base_ratio = min(prob * 1.2, 1.0)
        else:
            base_ratio = 0
            
        # 2Ï∞® ÌåêÎã®: RSI Í∏∞Î∞ò Î≥¥Ï†ï
        if rsi > 70:
            # RSI Í≥ºÎß§Ïàò(70Ïù¥ÏÉÅ) -> Í≥†Ï†ê ÏúÑÌóò -> ÎπÑÏ§ë 50% Ï∂ïÏÜå
            final_ratio = base_ratio * 0.5
        elif rsi < 30:
            # RSI Í≥ºÎß§ÎèÑ(30Ïù¥Ìïò) -> Î∞òÎì± Í∏∞Ìöå -> ÎπÑÏ§ë 1.5Î∞∞ ÌôïÎåÄ (ÏµúÎåÄ 100%)
            final_ratio = min(base_ratio * 1.5, 1.0)
            
            # (ÏòµÏÖò) Î™®Îç∏ ÌôïÏã†Ïù¥ Î∂ÄÏ°±Ìï¥ÎèÑ(0.4Ïù¥ÏÉÅ) RSIÍ∞Ä Îß§Ïö∞ ÎÇÆÏúºÎ©¥ ÏÜåÏï° ÏßÑÏûÖ
            if prob > 0.4 and base_ratio == 0:
                final_ratio = 0.3 
        else:
            final_ratio = base_ratio
            
        # Îß§Îß§ Î°úÏßÅ
        target_btc_value = portfolio_value * final_ratio
        current_btc_value = btc_holdings * current_price
        
        if target_btc_value > current_btc_value:
            buy_val = target_btc_value - current_btc_value
            if buy_val > cash: buy_val = cash
            if buy_val > 0:
                amount = (buy_val * (1 - transaction_fee)) / current_price
                btc_holdings += amount
                cash -= buy_val
        elif target_btc_value < current_btc_value:
            sell_val = current_btc_value - target_btc_value
            amount = sell_val / current_price
            if amount > btc_holdings: amount = btc_holdings
            if amount > 0:
                cash_gained = amount * current_price * (1 - transaction_fee)
                btc_holdings -= amount
                cash += cash_gained
                
    final_val = portfolio_values[-1]
    ret = (final_val - initial_capital) / initial_capital * 100
    return final_val, ret

# 3. Ïã§Ìñâ
final_capital, lv4_return = composite_strategy(my_prob, test_prices_aligned, test_dates_aligned, rsi_values)

print(f"Level 4 Composite Strategy Result")
print(f"Final Capital: ${final_capital:,.2f}")
print(f"Return: {lv4_return:.2f}%")
print(f"(vs Level 3: {my_result['total_return']:.2f}%)")
````

````python
# TODO: ÏûêÏã†ÎßåÏùò Î™®Îç∏ Íµ¨ÌòÑ

import torch.nn as nn
import torch.optim as optim
class MyTradingModel(nn.Module):
    """
    GRU Í∏∞Î∞òÏùò Ìä∏Î†àÏù¥Îî© Î™®Îç∏
    ÏûÖÎ†•: (batch, sequence_length, features)
    Ï∂úÎ†•: (batch, 1) - ÏÉÅÏäπ ÌôïÎ•† (0~1)
    """
    def __init__(self, input_size, hidden_size=128, num_layers=2, dropout=0.3):
        super(MyTradingModel, self).__init__()
        
        # 1. GRU Î†àÏù¥Ïñ¥ (LSTMÎ≥¥Îã§ ÌïôÏäµ ÏÜçÎèÑÍ∞Ä Îπ†Î¶Ñ)
        self.gru = nn.GRU(
            input_size=input_size, 
            hidden_size=hidden_size, 
            num_layers=num_layers, 
            batch_first=True, 
            dropout=dropout
        )
        
        # 2. Î∂ÑÎ•òÍ∏∞ (Classifier)
        self.fc = nn.Sequential(
            nn.Linear(hidden_size, 32),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(32, 1),
            nn.Sigmoid()  # 0~1 ÏÇ¨Ïù¥ ÌôïÎ•† Ï∂úÎ†•
        )
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_size)
        out, _ = self.gru(x)
        
        # ÎßàÏßÄÎßâ ÏãúÏ†ê(Last time step)Ïùò Hidden StateÎßå ÏÇ¨Ïö©
        # out shape: (batch_size, seq_len, hidden_size)
        last_step_out = out[:, -1, :] 
        
        # Î∂ÑÎ•òÍ∏∞ ÌÜµÍ≥º
        prediction = self.fc(last_step_out)
        return prediction
# Î™®Îç∏ ÏÉùÏÑ± (GPU ÏÇ¨Ïö©)
my_model = MyTradingModel(
    input_size=X_train_seq.shape[2],
    hidden_size=128, 
    num_layers=2,
    dropout=0.3
).to(device)
print("=== My GRU Model Created ===")
print(my_model)
print(f"\nTotal parameters: {sum(p.numel() for p in my_model.parameters()):,}")
````

````python
# TODO: ÏòàÏ∏° Î∞è Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò (Ï£ºÏÑù Ìï¥Ï†ú ÌõÑ ÏÇ¨Ïö©)

# 1. Î™®Îç∏ ÏòàÏ∏°
my_prob, my_pred = predict_with_probability(my_model, test_loader)

# 2. RSI Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (BTC FeaturesÏóêÏÑú Ï∂îÏ∂ú)
# test_start_idxÎäî Ïù¥Ï†Ñ ÏÖÄÎì§Ïóê Ï†ïÏùòÎêòÏñ¥ ÏûàÏäµÎãàÎã§.
rsi_values = btc_features['RSI_14'].iloc[test_start_idx:test_start_idx+len(my_prob)].values

# 3. Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò (Model + RSI Logic)
print("Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏßÑÌñâ Ï§ë...")
cash = 10000
btc_holdings = 0
portfolio_values = []
transaction_fee = 0.001
threshold = 0.65
# Í±∞Îûò ÌöüÏàò Ïπ¥Ïö¥Ìä∏
trade_count = 0 
for i in range(len(my_prob)):
    current_price = test_prices_aligned[i]
    prob = my_prob[i]
    rsi = rsi_values[i]
    
    portfolio_value = cash + btc_holdings * current_price
    portfolio_values.append(portfolio_value)
    
    # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
    if i == len(my_prob) - 1:
        if btc_holdings > 0:
            cash += btc_holdings * current_price * (1 - transaction_fee)
        continue
        
    # --- Ï†ÑÎûµ Î°úÏßÅ ÏãúÏûë ---
    # 1. Í∏∞Î≥∏ ÎπÑÏ§ë (Î™®Îç∏ ÌôïÎ•† Í∏∞Î∞ò)
    if prob > threshold:
        base_ratio = min(prob * 1.2, 1.0) # ÌôïÎ•†Ïùò 1.2Î∞∞ Ìà¨Ïûê (ÏµúÎåÄ 100%)
    else:
        base_ratio = 0
        
    # 2. RSI Î≥¥Ï†ï
    if rsi > 70: # Í≥ºÎß§Ïàò -> ÎπÑÏ§ë Ï∂ïÏÜå
        final_ratio = base_ratio * 0.5
    elif rsi < 30: # Í≥ºÎß§ÎèÑ -> ÎπÑÏ§ë ÌôïÎåÄ
        final_ratio = min(base_ratio * 1.5, 1.0)
        # Î™®Îç∏ ÌôïÏã†Ïù¥ Î∂ÄÏ°±Ìï¥ÎèÑ RSIÍ∞Ä ÎÇÆÏúºÎ©¥ ÏÜåÏï° ÏßÑÏûÖ
        if prob > 0.4 and base_ratio == 0:
            final_ratio = 0.3
    else:
        final_ratio = base_ratio
    # --- Ï†ÑÎûµ Î°úÏßÅ ÎÅù ---
        
    # Îß§Îß§ Ïã§Ìñâ
    target_btc_value = portfolio_value * final_ratio
    current_btc_value = btc_holdings * current_price
    
    if target_btc_value > current_btc_value: # Îß§Ïàò
        buy_val = target_btc_value - current_btc_value
        if buy_val > cash: buy_val = cash
        if buy_val > 0:
            amount = (buy_val * (1 - transaction_fee)) / current_price
            btc_holdings += amount
            cash -= buy_val
            trade_count += 1
            
    elif target_btc_value < current_btc_value: # Îß§ÎèÑ
        sell_val = current_btc_value - target_btc_value
        amount = sell_val / current_price
        if amount > btc_holdings: amount = btc_holdings
        if amount > 0:
            cash_gained = amount * current_price * (1 - transaction_fee)
            btc_holdings -= amount
            cash += cash_gained
            trade_count += 1
# 4. Í≤∞Í≥º Ï†ïÎ¶¨
my_result = {
    'initial_capital': 10000,
    'final_value': portfolio_values[-1],
    'total_return': (portfolio_values[-1] - 10000) / 10000 * 100,
    'portfolio_values': portfolio_values,
    'num_trades': trade_count,
    'total_fees_paid': 0 # Í≥ÑÏÇ∞ ÏÉùÎûµ
}
print("="*70)
print("ÎÇòÏùò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ Í≤∞Í≥º (Level 4 Composite)")
print("="*70)
print(f"Ï¥àÍ∏∞ ÏûêÎ≥∏: ${my_result['initial_capital']:,.2f}")
print(f"ÏµúÏ¢Ö ÏûêÎ≥∏: ${my_result['final_value']:,.2f}")
print(f"ÏàòÏùµÎ•†: {my_result['total_return']:.2f}%")
print(f"Buy and Hold ÎåÄÎπÑ: {my_result['total_return'] - buy_hold_return:.2f}%p")
print(f"Í±∞Îûò ÌöüÏàò: {my_result['num_trades']}Ìöå")
print("="*70)
````
